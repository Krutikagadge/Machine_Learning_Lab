```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['species']=pd.Categorical.from_codes(iris.target,iris.target_names)
df.head()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df3 = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df3['species'] = [iris.target_names[i] for i in iris.target]

print("Before Label Encoding:", df3['species'].head())

le = LabelEncoder()
df3['species_encoded'] = le.fit_transform(df3['species'])


print("After Label Encoding:", df3[['species', 'species_encoded']].head())
```

```python
arr = [100,120,130,150,180,200,220,250]
length = len(arr)
q1 = 0.25 * (length+1)
min,max = int(q1)-1,int(q1)
min_value = arr[min]
max_value = arr[max]
print(min_value,max_value)
q1_value = min_value + 0.25 * (max_value-min_value)
print(q1_value)
q3 = 0.75 * (length+1)
min3,max3 = int(q3)-1,int(q3)
min_value3 = arr[min3]
max_value3 = arr[max3]
print(min_value3,max_value3)
q3_value = min_value3 + 0.75 * (max_value3-min_value3)
print(q3_value)
iqr = q3_value-q1_value
print(iqr)
lb = q1_value - 1.5 * iqr
ub = q3_value + 1.5 * iqr
print(lb,ub)
```
